```python
def get_fields(reference):
    
    publisher = None
    
    elements = reference.split('.')
    
    authors = elements[0]

    title = re.search(r"\"(.+?)\"", reference)
    date = re.search(r".*([1-3][0-9]{3})", reference).group(1)
    rest = elements[1:]
        
    if title:
        title = title.group(1)[:-1]
        atd = [title, authors, date]
    else:
        atd = [authors, date]
    
    keep = []
    for e in elements:
        res = any(a in e for a in atd)
        if not res and e != '':
            keep.append(e)

        if date in e:
            if '(' in e:
                B = e.split('(')
                for b in B:
                    if date not in b:
                        keep.append(b)
            else:
                C = e.split(', ')
                for c in C:
                    if date not in c:
                        publisher = c

    if title:
        journal_or_book = keep[0][2:-1]
    else:
        journal_or_book = keep[0][1:]
    
    res_dict = {
        'title' : title,
        'authors' : authors,
        'date' : date,
        'journal_or_book' : journal_or_book,
        'publisher' : publisher
    }
    
    return(res_dict)
```

```python
reference = "García-Martín, Eva, Crefeda Faviola Rodrigues, Graham Riley, and Håkan Grahn. \"Estimation of energy consumption in machine learning.\" Journal of Parallel and Distributed Computing 134 (2019): 75-88"
res_dict = get_fields(reference)
print(res_dict)
```

```python
reference = "Zhou, Zhi-Hua. Machine learning. Springer Nature, 2021."
res_dict = get_fields(reference)
print(res_dict)
```

```python
reference = "Wu, Gene, Joseph L. Greathouse, Alexander Lyashevsky, Nuwan Jayasena, and Derek Chiou. \"GPGPU performance and power estimation using machine learning.\" In 2015 IEEE 21st international symposium on high performance computer architecture (HPCA), pp. 564-576. IEEE, 2015."
res_dict = get_fields(reference)
print(res_dict)
```
