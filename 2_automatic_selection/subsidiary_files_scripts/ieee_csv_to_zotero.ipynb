This file can be used to extract the doi (or other identifyier) from the ieee csv file and us them to import the results into zotero.

```python
import pandas as pd

_path = '..\saved_results_other_datasources\ieee\export2023.06.12-11.30.07.csv'

df = pd.DataFrame(pd.read_csv(_path))
```

```python
df.columns
```

We use the following lists of ISBNs and DOIs to import the corresponing papers into zotero.

```python
# papers without any DOI have ISBNs:
ISBNs = df[df['DOI'].isnull().values]['ISBNs'].to_list()
print(", ".join(ISBNs))
```

```python
DOIs = df[df['DOI'].isnull().values == False]['DOI'].tolist()
print(", ".join(DOIs))
```

```python
#look for duplicates with different DOIs:
test = df['Document Title'].tolist()

from collections import Counter
test_dict = dict(Counter(test))

for key in test_dict.keys():
    if test_dict[key]>1:
        print(key)
```
